```python
import pandas as pd
import numpy as np
from xgboost import XGBRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score
from matplotlib import pyplot
```

## Data Loading

```python
data = pd.read_csv('train.csv', header=0)
test = pd.read_csv('test.csv', header=0)
print(data[:5], data.shape, test.shape)
```

## Data Exploration

```python
data.isnull().values.any()
```

```python
for n in range(data.shape[1]):
    print(data.columns[n], ": ", data.iloc[:,n].unique())
```

## Data Preprocessing

```python
np.unique(np.append(data.iloc[:,2].unique(), test.iloc[:,1].unique()))
```

```python
label_encoder = {}
one_hot_encoder = {}
for n in [2,3,4,5,6,7,8,9]:
    print(n, ": ")
    
    #fit encoder
    label_unique = np.unique(np.append(data.iloc[:,n].unique(), test.iloc[:,n-1].unique()))
    label_encoder[n] = LabelEncoder().fit(label_unique)
    one_hot_encoder[n] = OneHotEncoder(sparse=False).fit(np.arange(len(label_encoder[n].classes_)).reshape(-1,1))
    
    #process train data
    data.iloc[:,n] = label_encoder[n].transform(data.iloc[:,n])
    features = one_hot_encoder[n].transform(data.iloc[:,n].values.reshape(-1,1))
    print(data.shape, ' + ', features.shape)
    data = pd.concat([data, pd.DataFrame(features)], axis=1)
    
    #process test data
    test.iloc[:,n-1] = label_encoder[n].transform(test.iloc[:,n-1])
    features = one_hot_encoder[n].transform(test.iloc[:,n-1].values.reshape(-1,1))
    print(test.shape, ' + ', features.shape)
    test = pd.concat([test, pd.DataFrame(features)], axis=1)
```

```python
print(data[:5], data.shape, test.shape)
```

## Train Test Split

```python
dataset = data.drop(data.columns[[0,2,3,4,5,6,7,8,9]], axis=1).values.astype(int)
X = dataset[:,[i for i in range(dataset.shape[1]) if i not in [0]]]
Y = dataset[:,0]
```

```python
seed = 420
test_size = 0.33
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=test_size, 
                                                    random_state=seed)
```

## Model Training

```python
model = XGBRegressor()
kfold = KFold(n_splits=10, random_state=420)
cv = cross_val_score(model, X_train, Y_train, cv=kfold)
```

```python
#model = XGBRegressor(n_jobs=-1, max_depth=4, learning_rate=0.0045, n_estimators=10000, 
                     #colsample_bylevel=0.93, base_score=Y_train.mean())
model = XGBRegressor(n_jobs=-1, max_depth=4, learning_rate=0.005, n_estimators=10000, 
                     subsample=0.8, colsample_bytree=0.8, base_score=Y_train.mean())
model.fit(X_train, Y_train, eval_metric=['rmse'], 
          eval_set=[(X_train, Y_train), (X_test, Y_test)],
         early_stopping_rounds=400)
```

```python
model2 = XGBRegressor(n_jobs=-1, max_depth=4, learning_rate=0.005, n_estimators=10000, 
                      subsample=0.8, colsample_bytree=0.8, base_score=Y_train.mean(), 
                      random_state=420)
model2.fit(X_train, Y_train, eval_metric=['rmse'], 
          eval_set=[(X_train, Y_train), (X_test, Y_test)],
         early_stopping_rounds=400)
```

```python
model3 = XGBRegressor(n_jobs=-1, max_depth=4, learning_rate=0.005, n_estimators=10000, 
                      subsample=0.8, colsample_bytree=0.8, base_score=Y_train.mean(), 
                      random_state=50)
model3.fit(X_train, Y_train, eval_metric=['rmse'], 
          eval_set=[(X_train, Y_train), (X_test, Y_test)],
         early_stopping_rounds=400)
```

## Model Performance

```python
print(cv)
print('CV Score: %.2f%% (%.2f%%)' % (cv.mean()*100, cv.std()*100))
```

```python
Y_pred = model.predict(X_test)
```

```python
score = r2_score(Y_test, Y_pred)
print("Score: %.2f%%" % (score*100.0))
```

```python
Y_pred2 = model2.predict(X_test)
```

```python
score2 = r2_score(Y_test, Y_pred2)
print("Score: %.2f%%" % (score2*100.0))
```

```python
Y_pred3 = model3.predict(X_test)
```

```python
score3 = r2_score(Y_test, Y_pred3)
print("Score: %.2f%%" % (score3*100.0))
```

```python
fig, ax = pyplot.subplots()
ax.plot(x_axis, eval_metrics['validation_0']['rmse'], label='Train')
ax.plot(x_axis, eval_metrics['validation_1']['rmse'], label='Test')
ax.legend()
pyplot.ylabel('RMSE')
pyplot.title('XGBoost RMSE')
pyplot.show()
```
